# Matplotlib Pyplot

```python
#import plplot submodule as alias plt
import matplotlib.pyplot as plt

import numpy as np
```

# Matplotlib Plotting
plot() function : parameter 1 is an array of points on the x-axis, parameter 2 is an array of points on the y-axis

```python

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

#line plot
plt.plot(xpoints, ypoints)
plt.show()
```

**default x-points**: If we do not specify the points on the x-axis, default x-points will be assigned 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
#default x-points
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

Markers: https://www.w3schools.com/python/matplotlib_markers.asp

```python
# use a square marker for points

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, marker='s')
plt.show()
```

**Format strings**
(shortcut string notation) : *marker*|*line*|*colour*

```python
# use a square marker for points, dotted line, red colour

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 's:r')
plt.show()
```

```python
# use a circle marker, market size = 20 , marker edge colour = red

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, marker='o', ms=20, mec='r')
plt.show()
```

# Line
**line style** : *linestyle* or *ls*,
**line colour** : *color* or *c*, **line width** : *linewidth* or *lw*


```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted', color='r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color='r', linewidth='20.5')
plt.show()
```

```python
plt.plot(ypoints, linestyle = 'dashed')
```

**Multiple lines**

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

You can also plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

# Labels

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

# Font properties
*fontdict* parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

# Multiple plots in one figure : subplot()

https://www.w3schools.com/python/matplotlib_subplot.asp

The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

```python
# figure with one row and two columns

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])


plt.subplot(1, 2, 1)  #the figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x,y)

plt.show()
```

```python
# figure with one column and two rows

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP") #supertitle
plt.show()
```

**Histogram : frequency distribution**
